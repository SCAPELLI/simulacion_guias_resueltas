## Ejercicio 22

Utilizando Octave/Matlab o Python, realice el siguiente experimento 100 veces: genere 50 variables
aleatorias independientes e idÃ©nticamente distribuidas con la siguiente distribuciÃ³n discreta de
probabilidades

Aplique un test de Chi2 con una distribuciÃ³n correspondiente a un dado no cargado con un nivel de significancia del 5%. 

Â¿Con quÃ© frecuencia se rechaza la hipÃ³tesis nula?
Â¿CÃ³mo se interpreta este resultado?


```python
import random
from scipy import stats
from collections import Counter

alpha = 0.05
zona_de_rechazo = 0
zona_de_aceptacion = 0
for i in range(100):
    l = []
    m = 50
    for i in range(m):
        a = random.choices([1, 2, 3, 4, 5, 6], [4, 4, 4, 4, 3, 5])
        l.append(a[0])

    ocurrencias_observadas = list(Counter(l).values()) 
    d2, pValor = stats.chisquare(ocurrencias_observadas, m * 1/6)
    if pValor >= alpha:
        zona_de_aceptacion += 1
    else:
        zona_de_rechazo += 1
print(f"muestras que caen en zona de aceptacion: {zona_de_aceptacion}")
print(f"muestras que caen en zona de rechazo   : {zona_de_rechazo}")
```

```python
import numpy as np
import pandas as pd

aceptados = 0
aceptados_2 = 0
rechazados = 0
rechazados_2 = 0
for i in range(500):
    #d)
    arreglo_de_X1 = []
    m = 100
    for i in range(m):
        x = 0
        for i in range(10):
            x += np.random.uniform()
        arreglo_de_X1.append(x)

    #e)
    arreglo_de_X2 = []
    for i in range(m):
        x = 0
        for i in range(20):
            x += np.random.uniform()
        arreglo_de_X2.append(x)

#     import seaborn as sns
#     sns.kdeplot(arreglo_de_X1)
#     sns.kdeplot(arreglo_de_X2)

    #f)

    # estudiamos la esperanza y varianza de X (item d) (planteado en cuaderno)
    e1 = 5
    v1 = 5/6

    # estudiamos la esperanza y varianza de X (item f) (idem item d)
    e2 = 10
    v2 = 5/3

    stats.norm.cdf(float('inf'))               # EJEMPLO DE USO DE FUNCION ACUMULADA F(X) DE LA normal estandar 

    # ----------------------------CALCULO DE PARAMETROS PARA TEST DE CHI2-------------------
    alpha = 0.05 
    clases = [float('-inf'),-0.8, -0.4, -0.2, 0, 0.2, 0.4, 0.8, float('inf')]                        # se definen los intervalos 

    ocurrencias_esperadas = []
    for i in range(1, len(clases)):
        probabilidad_del_intervalo = stats.norm.cdf(clases[i]) - stats.norm.cdf(clases[i-1])         # se calcula las ocurrencias para cada intervalo siendo que hay m muestras
        ocurrencias_esperadas.append(probabilidad_del_intervalo*m)
    
    def obtener_arreglo_normalizado(arreglo, e, v):                                                  # se usaran los valores de la muestra normalizados por esta funcion
        normalizado = []
        for i in arreglo:
            normalizado.append((i - e)/math.sqrt(v))
        return normalizado
    
    # import pandas as pd
    arreglo_normalizado_1 = obtener_arreglo_normalizado(arreglo_de_X1, e1, v1)
    ocurrencias_observadas_1 = pd.cut(arreglo_normalizado_1, clases).value_counts()                       # se cuentan las ocurrencias observadas en la muestra con magia de pandas 

    d2, pValor = stats.chisquare(ocurrencias_observadas_1, ocurrencias_esperadas)                       # finalmente se calcula el valor de d2 y pvalor con scipy.stats

    if pValor >= alpha:
        aceptados += 1
    else:
        rechazados += 1
    
    arreglo_normalizado_2 = obtener_arreglo_normalizado(arreglo_de_X2, e2, v2)
    ocurrencias_observadas_2 = pd.cut(arreglo_normalizado_2, clases).value_counts()                       # se cuentan las ocurrencias observadas en la muestra con magia de pandas 

    d2, pValor = stats.chisquare(ocurrencias_observadas_2, ocurrencias_esperadas)                       # finalmente se calcula el valor de d2 y pvalor con scipy.stats

    if pValor >= alpha:
        aceptados_2 += 1
    else:
        rechazados_2 += 1
    
print(f"Test chi2 aceptados item d: {aceptados}")
print(f"Test chi2 rechazados item d: {rechazados}")

print(f"Test chi2 aceptados item e: {aceptados_2}")
print(f"Test chi2 rechazados item e: {rechazados_2}")
```

```python
m = 100
pares = []
for i in range(m):
    x = np.random.uniform()
    y = np.random.uniform()
    pares.append((x, y))

# se divide el espacio en 100 celdas de 0.1x0.1
probabilidad_que_caiga_en_una_celda = 0.1 * 0.1
ocurrencias_esperadas = m * probabilidad_que_caiga_en_una_celda

ocurrencias_observadas = np.zeros((m,m))
for muestra in pares:
    x = int(muestra[0] * 10)
    y = int(muestra[1] * 10)
    ocurrencias_observadas[x][y] += 1

d2, pValor = stats.chisquare(ocurrencias_observadas.flatten(), ocurrencias_esperadas)
print(f"pValor = {pValor}")
```

```python
def laplace(x):
    lamb = 1
    esperanza = 5
    return (lamb/2) * math.exp(-lamb*abs(x-esperanza))

def laplace_inversa(u):
    lamb = 10
    esperanza = 5
    return esperanza - (1/lamb) * (np.sign(u-0.5) * np.log(1-2*abs(u-0.5)))
m = 10000
x = np.arange(0, 10.1, 0.1)
plt.title("Funcion de densidad laplace")
plt.axhline(color="black")
plt.axvline(color="black")
plt.plot(x, [laplace(i) for i in x])
plt.show()

plt.title("histograma de nros generados")
nros_generados = [laplace_inversa(i) for i in np.random.uniform(size=m)]
plt.hist(nros_generados, bins = 100)
plt.show()
```

```python
m = 1
alpha = 1.5
def pareto(x):
    return pow(alpha*m, alpha)/pow(x, alpha+1)
def pareto_inversa(u):
    return m/pow(1-u, alpha)
x = np.arange(1, 10, 0.1)
plt.plot(x, [pareto(i) for i in x])
plt.show()

nros_generados = [pareto_inversa(i) for i in np.random.uniform(size=10000)]
plt.hist(nros_generados, bins=100)
plt.show()
```

```python
lamb = 1
alpha = 1
def weibull(x):
    return lamb*alpha*pow(lamb*alpha, alpha-1)*math.exp(-pow(lamb*x,alpha))
def weibull_inversa(u):
    return pow(-np.log(1-u), alpha)/lamb

x = np.arange(0.1, 10, 0.1)
plt.plot(x, [weibull(i) for i in x])
plt.show()

nros_generados = [weibull_inversa(i) for i in np.random.uniform(size=1000)]
plt.hist(nros_generados)
plt.show()
```

```python
# Para la demostracion hay que aplicar el metodo de aceptacion rechazo de tipo 2 usando un generador de nros con dist. exponencial
import random

def Z():
    seguir = True
    while seguir:
        u1 = np.random.uniform()
        x1 = -np.log(u1)
        u2 = np.random.uniform()
        if u2<=math.exp(-pow(x1-1,2)/2):
            seguir = False
    sgn = random.choices([1, -1], [1/2, 1/2])
    return sgn[0] * x1

nros_generados = [Z() for i in range(1000)]
plt.hist(nros_generados)
plt.show()
print(f"esperanza: {np.mean(nros_generados)}")
print(f"varianza: {np.var(nros_generados)}")
```
